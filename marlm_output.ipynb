```python
# -*- coding: utf-8 -*-
"""
Created on Wed May 15 11:13:36 2019

@author: kempf
"""
import pandas as pd
import glob
import numpy as np
from tkinter import *
from matplotlib.backends.backend_tkagg import (
    FigureCanvasTkAgg, NavigationToolbar2Tk)
import matplotlib.pyplot as plt
import matplotlib
import xlrd
import time
from datetime import datetime
matplotlib.use("TkAgg")
#from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg, NavigationToolbar2TkAgg
from matplotlib.figure import Figure
import tkinter.messagebox
print(time.strftime("%d.%m.%Y %H:%M:%S"))

# Einlesen der deutschen CSV-Dateien
path = r'\\SRV-FILE01\Studierende\Kempf\Strompreis-Börsen-Plot\Test' #Dateipfad
allFiles = glob.glob(path + "/*.csv") #Alle CSV-Datein im path in allFiles speichern
Fertig = {}
for file_ in allFiles: #Alle CSV durchgehen und die Werte in list_, bzw. df ablegen
    df = pd.read_csv(file_,index_col=None, header=1, sep=',',
         names=["Datum",
                "Woche",
                "Wochentag",
                "Stunde",
                "Preis",
                "Volumen",
                "Kauf/Verkauf"], engine='python')
    #list_.append(df)
    data_dict = df.to_dict() #DataFrame zu Dictionary umwandeln
    data_dict.pop('Woche', None) #Woche unnötig, ist doch über Datum eindeutig festgelegt
    data_dict.pop('Wochentag', None) #Siehe Woche
    Zwischenvariable = data_dict['Datum'] #Datum auslesen
    Datum = Zwischenvariable[0] #Ersten Wert nehmen (Immer das gleiche Datum)
    data_dict.pop('Datum', None) #Datum entfernen
    Fertig [Datum] = data_dict #Datum als Key einfügen mit untergeordnetem Dictionary


# Einlesen der Niederländischen Excel-Dateien
print(time.strftime("%d.%m.%Y %H:%M:%S"))
path = r'\\SRV-FILE01\Studierende\Kempf\Strompreis-Börsen-Plot\Test_Excel' #Dateipfad
allFiles3 = glob.glob(path + "/*.xls") #Alle Excel-Datein im path in allFiles speichern
Fertig3 = {}
for file_3 in allFiles3: #Alle Excel-Files durchgehen und die Werte in list_, bzw. df ablegen  
    df3 = pd.read_excel(file_3,
                names=["Datum",
                "Stunde",
                "Kauf/Verkauf",
                "Volumen",
                "Preis",
                "Ordertype"])
    Zwischendict = {} #Zwischenspeicherdictionary, ist nötig, weil jede Stunde eine eigene Datei hat  
    
    data_dict3 = df3.to_dict() #DataFrame zu Dictionary umwandeln
    data_dict3.pop('Ordertype', None) #Ordertyp unnötig
    Zwischenvariable3 = data_dict3['Datum'] #Datum auslesen
    Datum3 = Zwischenvariable3[0] #Ersten Wert nehmen (Immer das gleiche Datum)
    Datum3 = Datum3.strftime("%d/%m/%Y") #Timestamp in TT/MM/JJJJ-Format umwandeln
    data_dict3.pop('Datum', None) #Datum entfernen
    Zwischenvariable3eins = data_dict3['Stunde'] #Stunde zwischenspeichern
    Stunde3 = Zwischenvariable3eins[0] #Aktuell zu bearbeitende Stunde
    data_dict3.pop('Stunde', None) #Stunde entfernen
    
    Zwischendict [Stunde3] = data_dict3 #Zwischenspeicherdictionary mit den Daten der eingelesen Datei anlegen
    
    #Fertig3  = Datum3 #Datum als Key einfügen mit untergeordnetem Dictionary 
    #if-Bedingung notwendig, weil bei neuem Tag erst ein Key in Fertig3 angelegt werden muss
    if Datum3 not in Fertig3:
        Fertig3 [Datum3] = Zwischendict
    else:
        Fertig3[Datum3].update(Zwischendict)
    
    


global Anzahl_Show_gedrueckt
Anzahl_Show_gedrueckt = 0  
def Verschieben(): 
             if Anzahl_Show_gedrueckt == 0:
                 tkFensterFehler = Toplevel()
                 tkFensterFehler.title('Fehler')
                 tkFensterFehler.geometry('400x30')
                 labelFehler = Label(master=tkFensterFehler, text='Zuerst mindestens einmal Show drücken!!!')
                 labelFehler.place(x=10, y=10, width=380, height=20)
             else:
                 canvas = FigureCanvasTkAgg(fig, master = master)
                 canvas._tkcanvas.grid(row = 3)
                 axis = fig.add_subplot(111)
       
                 canvas2 = FigureCanvasTkAgg(fig2, master = master)
                 canvas2._tkcanvas.grid(row = 3, column=2)
                 axis2 = fig2.add_subplot(111)
                 
                 global yExcel
                 global yCSV
                 global xSell
                 global ySell
                 global xBuy
                 global yBuy
                   
                 global xSell2
                 global ySell2
                 global xBuy2
                 global yBuy2
                 einmal_gelaufen = 0
                  
                 while yExcel[0] > yCSV[0]:
                     
                     
                     if einmal_gelaufen == 0:
                         xBuy_Verschoben = [x+10 for x in xBuy]
                         xSell2_Verschoben = [x+10 for x in xSell2]
                         
                     else:
                         xBuy_Verschoben = [x+10 for x in xBuy_Verschoben]
                         xSell2_Verschoben = [x+10 for x in xSell2_Verschoben]
                     
                     xBuyArray = np.array(xBuy_Verschoben)
                     xSell2Array = np.array(xSell2_Verschoben)
                     
                     xSellArray = np.array(xSell)
                     ySellArray = np.array(ySell)
                     yBuyArray = np.array(yBuy)
                   
                     ySell2Array = np.array(ySell2)
                     xBuy2Array = np.array(xBuy2)
                     yBuy2Array = np.array(yBuy2)
                     
                     xCSV, yCSV = intersection(xSellArray, ySellArray, xBuyArray, yBuyArray)
                     xExcel, yExcel = intersection(xSell2Array, ySell2Array, xBuy2Array, yBuy2Array)
                     einmal_gelaufen = einmal_gelaufen + 1
                     
                 
                 
                 axis.plot(xBuy_Verschoben, yBuy, "--")
                 axis2.plot(xSell2_Verschoben,ySell2, "--")
                 axis.plot(xCSV, yCSV, 'go') 
                 axis2.plot(xExcel, yExcel, 'go')
                    
                 
                 toolbarFrame = Frame(master=master)
                 toolbarFrame.grid(row=10,column=0)
                 toolbar = NavigationToolbar2Tk(canvas, toolbarFrame)
                 toolbar.update()
       
                 toolbarFrame2 = Frame(master=master)
                 toolbarFrame2.grid(row=10,column=2)
                 toolbar2 = NavigationToolbar2Tk(canvas2, toolbarFrame2)
                 toolbar2.update()
        
def Schnittpunkt():
             if Anzahl_Show_gedrueckt == 0:
                 tkFensterFehler = Toplevel()
                 tkFensterFehler.title('Fehler')
                 tkFensterFehler.geometry('400x30')
                 labelFehler = Label(master=tkFensterFehler, text='Zuerst mindestens einmal Show drücken!!!')
                 labelFehler.place(x=10, y=10, width=380, height=20)
             else:
                 canvas = FigureCanvasTkAgg(fig, master = master)
                 canvas._tkcanvas.grid(row = 3)
                 axis = fig.add_subplot(111)
       
                 canvas2 = FigureCanvasTkAgg(fig2, master = master)
                 canvas2._tkcanvas.grid(row = 3, column=2)
                 axis2 = fig2.add_subplot(111)
                 
                 axis.annotate( [xCSV[0], yCSV[0]],
                 xy=(xCSV[0], yCSV[0]), xycoords='data',
                 xytext=(-50, -20), textcoords='offset points', fontsize=10)
                 
                 axis2.annotate( [xExcel[0], yExcel[0]],
                 xy=(xExcel[0], yExcel[0]), xycoords='data',
                 xytext=(-50, -20), textcoords='offset points', fontsize=10)
                 
                 toolbarFrame = Frame(master=master)
                 toolbarFrame.grid(row=10,column=0)
                 toolbar = NavigationToolbar2Tk(canvas, toolbarFrame)
                 toolbar.update()
       
                 toolbarFrame2 = Frame(master=master)
                 toolbarFrame2.grid(row=10,column=2)
                 toolbar2 = NavigationToolbar2Tk(canvas2, toolbarFrame2)
                 toolbar2.update()
                 
             
def buttonOkClick(): #Fenster schließen
            tkFensterFehler.quit()
            
#Schnittpunktbestimmung der Graphen (Fertig aus dem Internet verwendet)            
def intersection(x1,y1,x2,y2):
    """
INTERSECTIONS Intersections of curves.
   Computes the (x,y) locations where two curves intersect.  The curves
   can be broken with NaNs or have vertical segments.
usage:
x,y=intersection(x1,y1,x2,y2)
    Example:
    a, b = 1, 2
    phi = np.linspace(3, 10, 100)
    x1 = a*phi - b*np.sin(phi)
    y1 = a - b*np.cos(phi)
    x2=phi    
    y2=np.sin(phi)+2
    x,y=intersection(x1,y1,x2,y2)
    plt.plot(x1,y1,c='r')
    plt.plot(x2,y2,c='g')
    plt.plot(x,y,'*k')
    plt.show()
    """
    ii,jj=_rectangle_intersection_(x1,y1,x2,y2)
    n=len(ii)

    dxy1=np.diff(np.c_[x1,y1],axis=0)
    dxy2=np.diff(np.c_[x2,y2],axis=0)

    T=np.zeros((4,n))
    AA=np.zeros((4,4,n))
    AA[0:2,2,:]=-1
    AA[2:4,3,:]=-1
    AA[0::2,0,:]=dxy1[ii,:].T
    AA[1::2,1,:]=dxy2[jj,:].T

    BB=np.zeros((4,n))
    BB[0,:]=-x1[ii].ravel()
    BB[1,:]=-x2[jj].ravel()
    BB[2,:]=-y1[ii].ravel()
    BB[3,:]=-y2[jj].ravel()

    for i in range(n):
        try:
            T[:,i]=np.linalg.solve(AA[:,:,i],BB[:,i])
        except:
            T[:,i]=np.NaN


    in_range= (T[0,:] >=0) & (T[1,:] >=0) & (T[0,:] <=1) & (T[1,:] <=1)

    xy0=T[2:,in_range]
    xy0=xy0.T
    return xy0[:,0],xy0[:,1]

def _rect_inter_inner(x1,x2):
    n1=x1.shape[0]-1
    n2=x2.shape[0]-1
    X1=np.c_[x1[:-1],x1[1:]]
    X2=np.c_[x2[:-1],x2[1:]]    
    S1=np.tile(X1.min(axis=1),(n2,1)).T
    S2=np.tile(X2.max(axis=1),(n1,1))
    S3=np.tile(X1.max(axis=1),(n2,1)).T
    S4=np.tile(X2.min(axis=1),(n1,1))
    return S1,S2,S3,S4

def _rectangle_intersection_(x1,y1,x2,y2):
    S1,S2,S3,S4=_rect_inter_inner(x1,x2)
    S5,S6,S7,S8=_rect_inter_inner(y1,y2)

    C1=np.less_equal(S1,S2)
    C2=np.greater_equal(S3,S4)
    C3=np.less_equal(S5,S6)
    C4=np.greater_equal(S7,S8)

    ii,jj=np.nonzero(C1 & C2 & C3 & C4)
    return ii,jj





  
    
#Reaktion bei drücken der Taste Show
def show_entry_fields():
   fig.clf() #Figures für neuen Plot clearen
   fig2.clf() 
   
   
   
   user_selected_date = e1.get() #Usereingabe im Datumsfeld in Variable abspeichern
   user_selected_hour_raw = e2.get() #Usereingabe im Stundenfeld in Variable abspeichern
   user_selected_hour = int(user_selected_hour_raw) #Stunde in Integer-Wert umwandeln

# Fehlerhafte Stundeneingabe abfangen
   if user_selected_hour < 1 or user_selected_hour > 24:
       tkFensterFehler = Toplevel()
       tkFensterFehler.title('Fehler')
       tkFensterFehler.geometry('160x30')
       labelFehler = Label(master=tkFensterFehler, text='Keine sinnvolle Eingabe!')
       labelFehler.place(x=5, y=5, width=150, height=20)
# Fehlerhafte Datumseingabe abfangen
   if user_selected_date not in Fertig:
       tkFensterFehler = Toplevel()
       tkFensterFehler.title('Fehler')
       tkFensterFehler.geometry('160x30')
       labelFehler = Label(master=tkFensterFehler, text='Tag nicht in der Datenbank!')
       labelFehler.place(x=5, y=5, width=150, height=20)
    
   num_bids = len(Fertig[user_selected_date]['Stunde']) #Anzahl der Strombörsengebote auslesen
   
   
   #Bestimmung, wieviele Gebote in einer Stunde abgeben wurden und wieviele davon Kauf oder Verkaufsangebot waren
   i = 0 #Schleifenvariable
   bid_hour = 0 #Schleifenvariable
   
   global xSell
   global ySell
   global xBuy
   global yBuy
   
   global xSell2
   global ySell2
   global xBuy2
   global yBuy2
   
#Anfang und Ende der gewählten Stunde bestimmen  
   while bid_hour < user_selected_hour + 1 and i < num_bids - 1:
        bid_hour = Fertig[user_selected_date]['Stunde'][i]
        if user_selected_hour == 1:
            Anfang = 0
        if bid_hour < user_selected_hour:
            Anfang = i+1
        i = i+1
        
   Ende = i-2
   j = Anfang
 #Wechsel von Kauf zu Verkauf bestimmen   
   while j < Ende:
        if Fertig[user_selected_date]['Kauf/Verkauf'][j] == 'Purchase':
            break
        j = j+1
    
   Mitte =  j-1
   k = Anfang
   xSell = [] 
   #Liste der Verkaufswerte (Volumen)
   while k < Mitte:
        xSell.append(Fertig[user_selected_date]['Volumen'][k])
        k = k+1
   ySell = []
   k = Anfang
   #Liste der Verkaufswerte (Preis)
   while k < Mitte:
        ySell.append(Fertig[user_selected_date]['Preis'][k])
        k = k+1
    
   k = Mitte+1
   xBuy = []
   #Liste der Kaufswerte (Volumen)
   while k <= Ende:
        xBuy.append(Fertig[user_selected_date]['Volumen'][k])
        k = k+1
   yBuy = []
   k = Mitte+1
   #Liste der Kaufswerte (Preis)
   while k <= Ende:
        yBuy.append(Fertig[user_selected_date]['Preis'][k])
        k = k+1
   
   
    
    
    
    
# Gleiches Prozederewie bei CSV-Dateien nur auf Excel-Files angepasst
   num_bids2 = len(Fertig3[user_selected_date][user_selected_hour]['Volumen']) 
   m = 0
   #Wechsel von Kauf zu Verkauf bestimmen    
   while Fertig3[user_selected_date][user_selected_hour]['Kauf/Verkauf'][m]=='P':
       m = m+1
   m = m-1 
   xBuy2 = []
   n = 0
   #Liste der Kaufswerte (Volumen)
   while n <= m:
        xBuy2.append(Fertig3[user_selected_date][user_selected_hour]['Volumen'][n])
        n = n+1
   yBuy2 = []
   n = 0
   #Liste der Kaufswerte (Preis)
   while n <= m:
        yBuy2.append(Fertig3[user_selected_date][user_selected_hour]['Preis'][n])
        n = n+1
        
   m = m+1
   Speicher = m
   ySell2 = []
   #Liste der Verkaufswerte (Volumen)
   while m <= num_bids2 - 1:
        ySell2.append(Fertig3[user_selected_date][user_selected_hour]['Preis'][m])
        m = m+1
   xSell2 = []
   m = Speicher
   #Liste der Verkaufswerte (Preis)
   while m <= num_bids2 - 1:
        xSell2.append(Fertig3[user_selected_date][user_selected_hour]['Volumen'][m])
        m = m+1   
        
    
# Fenster konfigurieren und Plots einfügen
   canvas = FigureCanvasTkAgg(fig, master = master)
   canvas._tkcanvas.grid(row = 3)
   axis = fig.add_subplot(111)
   
   canvas2 = FigureCanvasTkAgg(fig2, master = master)
   canvas2._tkcanvas.grid(row = 3, column=2)
   axis2 = fig2.add_subplot(111)
   
# Kauf und Verkaufswerte plotten    
   axis.plot(xSell,ySell)
   axis.plot(xBuy, yBuy)
   axis2.plot(xSell2,ySell2)
   axis2.plot(xBuy2, yBuy2)
   
  
   
   #Kauf und Verkaufswerte in Arrays umwandeln
   xSellArray = np.array(xSell)
   ySellArray = np.array(ySell)
   xBuyArray = np.array(xBuy)
   yBuyArray = np.array(yBuy)
   
   xSell2Array = np.array(xSell2)
   ySell2Array = np.array(ySell2)
   xBuy2Array = np.array(xBuy2)
   yBuy2Array = np.array(yBuy2)
   
   #Schnittpunkt bestimmen
   global xCSV
   global yCSV
   xCSV, yCSV = intersection(xSellArray, ySellArray, xBuyArray, yBuyArray)
   axis.plot(xCSV, yCSV, 'bo')
   
   
#   x.astype(np.int64)
#   y.astype(np.int64)
#   np.round(x, 0)
#   np.round(y, 2)
 
    #Schnittpunkt mit Pfeil im Diagramm versehen
  
   #Schnittpunkt des zweiten Graphens
   global xExcel
   global yExcel
   xExcel, yExcel = intersection(xSell2Array, ySell2Array, xBuy2Array, yBuy2Array)
   axis2.plot(xExcel, yExcel, 'bo')
   
  # x.astype(np.int)
  # y.astype(np.int)
   #Schnittpunkt mit Pfeil im Diagramm versehen
   
   
 # x und y-Achsen beschriften
   axis.set_xlabel('Energiemenge [GWh]')
   axis.set_ylabel('Preis [€]')
   
   axis2.set_xlabel('Energiemenge [GWh]')
   axis2.set_ylabel('Preis [€]')
   
   axis.set_title('Deutschland')
   axis2.set_title('Niederlande')
# Gitter der Plots einschalten   
   axis.grid(True)
   axis2.grid(True)
#Funktionstoolbar aktualisieren   
   toolbarFrame = Frame(master=master)
   toolbarFrame.grid(row=10,column=0)
   toolbar = NavigationToolbar2Tk(canvas, toolbarFrame)
   toolbar.update()
   
   toolbarFrame2 = Frame(master=master)
   toolbarFrame2.grid(row=10,column=2)
   toolbar2 = NavigationToolbar2Tk(canvas2, toolbarFrame2)
   toolbar2.update()
   global Anzahl_Show_gedrueckt
   Anzahl_Show_gedrueckt = Anzahl_Show_gedrueckt+1
   
   

# Fenster definieren  
master = Tk()
Label(master, text="Datum (TT/MM/JJJJ)").grid(row=0) #Text zum Datum
Label(master, text="Stunde (1-24)").grid(row=1) #Text zur Stunde

Tk.wm_title(master, "Day-Ahead-Preise") #Titel des Fensters

e1 = Entry(master) #Datumseingabe
e2 = Entry(master) #Stundeneingabe

e1.insert(10,"31/12/2017") #Dummy-Eingaben
e2.insert(10,"1") #Dummy-Eingaben

e1.grid(row=0, column=1) #Ort des Eingabefeldes festlegen
e2.grid(row=1, column=1) #Ort des Eingabefeldes festlegen

Button(master, text='Show', command=show_entry_fields).grid(row=3, column=1, sticky=N, pady=4) #Button show definieren
Button(master, text='Schnittpunkt', command=Schnittpunkt).grid(row=3, column=1, sticky=W, pady=4) #Button Schnittpunkt definieren
Button(master, text='Verschieben', command=Verschieben).grid(row=3, column=1, sticky=S, pady=4) #Button Verschieben definieren

fig = Figure(figsize=(8,4), dpi=100) #Figure festlegen
axis = fig.add_subplot(111) #Plot festlegen

fig2 = Figure(figsize=(7,4), dpi=100) #Figure festlegen
axis2 = fig2.add_subplot(111) #Plot festlegen
#axis.plot(xSell,ySell)
#axis.plot(xBuy,yBuy)

#Achsenbeschriftung
axis.set_xlabel('Energiemenge [GWh]')
axis.set_ylabel('Preis [€]')
axis2.set_xlabel('Energiemenge [GWh]')
axis2.set_ylabel('Preis [€]')
axis.set_title('Deutschland')
axis2.set_title('Niederlande')
axis.grid(True)
axis2.grid(True)

# Plotfestlegung
canvas = FigureCanvasTkAgg(fig, master = master)
canvas._tkcanvas.grid(row = 3)
canvas2 = FigureCanvasTkAgg(fig2, master = master)
canvas2._tkcanvas.grid(row = 3, column=2)


#Toolbar einfügen
toolbarFrame = Frame(master=master)
toolbarFrame.grid(row=10,column=0)
toolbar = NavigationToolbar2Tk(canvas, toolbarFrame)
toolbar.update()

toolbarFrame2 = Frame(master=master)
toolbarFrame2.grid(row=10,column=2)
toolbar2 = NavigationToolbar2Tk(canvas2, toolbarFrame2)
toolbar2.update()

#Mainloop wartet auf Drücken bes Buttons
mainloop()
```
